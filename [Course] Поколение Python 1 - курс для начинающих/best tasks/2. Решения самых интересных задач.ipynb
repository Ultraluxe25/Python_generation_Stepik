### Решение шифра цезаря 2022 года

```python
low_en = 'abcdefghijklmnopqrstuvwxyz'
up_en = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

message = input().split()
cipher = ''  # шифр
# функция подсчёта количества букв в слове
for part in message:
    def find_length(word):
        counter = 0
        for c in word:
            if c.isalpha():
                counter += 1
        return counter
    
    for element in part:
        if element in low_en:
            cipher += low_en[(low_en.find(element) + find_length(part)) % 26]
        elif element in up_en:
            cipher += up_en[(up_en.find(element) + find_length(part)) % 26]
        else:
            cipher += element
    
    cipher += ' '  # отделяем новые слова пробелом
    
print(cipher[:-1])  # пробел в конце не включаем
```

### Решение шифра цезаря 2025 года (спустя 3.5 года)

```python
from string import ascii_lowercase, ascii_uppercase


def find_length(word: str) -> int:
    """Count symbols from alphabet.

    Args:
        word (str): each word of the text.

    Returns:
        int: number of non-digit and non-symbols (only text chars).    
    """
    return sum(char.isalpha() for char in word)


def swap_message(message: str) -> str:
    """Encrypt text in Caesar's cipher.

    Args:
        message (str): text before encryption.

    Returns:
        str: encrypted text
    """
    cipher = ''
    for part in message.split():
        for element in part:
            if element in ascii_lowercase:
                cipher += ascii_lowercase[(ascii_lowercase.find(element) + find_length(part)) % 26]
            elif element in ascii_uppercase:
                cipher += ascii_uppercase[(ascii_uppercase.find(element) + find_length(part)) % 26]
            else:
                cipher += element
        
        cipher += ' '  # Separate words in text

    return cipher.rstrip()


text = input()
print(swap_message(text))
```

```python
def print_salary(message: str):
    """Сalculate old and new salaries for all symbols in the message.
    
    Args:
        message (str): user's message.
        
    Returns:
        str: text of the messages and salaries for convertion.
    """
    # https://stackoverflow.com/questions/62871336/english-keyboard-text-to-russian-in-python
    eng_rus_common: dict[str] = dict(zip('eyopaxcETOPAHXCBM', 'еуорахсЕТОРАНХСВМ'))
    
    old_salary: int = 0
    additional_salary: int = 0
    
    for char in message:
        old_salary += ord(char) * 3
        if char in eng_rus_common:
            additional_salary += (ord(eng_rus_common[char]) - ord(char)) * 3 
        
    print(f"Старая стоимость: {old_salary}🐝")
    print(f"Новая стоимость: {old_salary + additional_salary}🐝")


text = input()
print_salary(text)
```
