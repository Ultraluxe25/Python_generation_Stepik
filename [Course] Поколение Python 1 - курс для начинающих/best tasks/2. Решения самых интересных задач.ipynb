```python
low_en = 'abcdefghijklmnopqrstuvwxyz'
up_en = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'

message = input().split()
cipher = ''  # шифр
# функция подсчёта количества букв в слове
for part in message:
    def find_length(word):
        counter = 0
        for c in word:
            if c.isalpha():
                counter += 1
        return counter
    
    for element in part:
        if element in low_en:
            cipher += low_en[(low_en.find(element) + find_length(part)) % 26]
        elif element in up_en:
            cipher += up_en[(up_en.find(element) + find_length(part)) % 26]
        else:
            cipher += element
    
    cipher += ' '  # отделяем новые слова пробелом
    
print(cipher[:-1])  # пробел в конце не включаем
```

```python
# put your python code here
def print_salary(message: str):
    """Сalculate salary for all symbols in the message.
    
    Args:
        message (str): user's message.
        
    Returns:
        str: text of the message and salary for convertion.
    """
    salary = 0
    for char in message:
        salary += ord(char) * 3
        
    print(f"Текст сообщения: '{message}'")
    print(f"Стоимость сообщения: {salary}🐝")


text = input()
print_salary(text)
```
