## Problem 1

```Python
def find_min_max_str():
    """Find min and max strings among inputs"""
    strings: list[str] = []
    
    while True:
        value: str = input()
        if value == 'КОНЕЦ':
            break
        
        strings.append(value)
    
    if strings:  # List might be empty
        print(f'Минимальная строка ⬇️: {min(strings)}')
        print(f'Максимальная строка ⬆️: {max(strings)}')

    
find_min_max_str()
```

## Problem 3




```Python
def check_class(checks_number: int):
    """Check whether class title is correct
    
    Args:
        checks_number (int): number of checks
    """
    for _ in range(checks_number):
        class_num = input()
        if len(class_num) == 2 and class_num[0].isdigit() \
        and ord(class_num[1]) >= 1040 and ord(class_num[1]) <= 1055:
            print('YES')
        else:
            print('NO')
        
        
# print(ord('П'), ord('А'))
check_class(int(input()))
```

```Python
# put your python code here
word_1 = input().lower()
word_2 = input().lower()

new_word_1 = ''
new_word_2 = ''

for i in word_1:
    if i.isalpha():
        new_word_1 += i
        
for i in word_2:
    if i.isalpha():
        new_word_2 += i
        
# print(new_word_1, new_word_2)  
print('YES' if new_word_1 == new_word_2 else 'NO')
```
