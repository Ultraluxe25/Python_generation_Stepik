# Task 2




```Python
def check_number(car_number: str) -> str:
    if not (car_number[1:4] + car_number[7:]).isdigit():
        return 'NO'
    
    chars = car_number[0] + car_number[4:6]
    for i in chars:
        if i not in '':
            return 'NO'
        
    if car_number[6] != '_' or len(car_number) > 10 or len(car_number) < 9:
        return 'NO'
    
    return 'YES'


# <><><><><><>_<><>
# <><><><><><>_<><><>
number = input()
print(check_number(number))
```

# Task 3




```Python
def check_user(nick_name: str) -> str:    
    if nick_name[0] != '@':
        return 'Incorrect'
    elif len(nick_name) < 5 or len(nick_name) > 15:
        return 'Incorrect'
    elif nick_name[1:].isalnum() and nick_name[1:] == nick_name[1:].lower():
        return 'Correct'
    
    return 'Incorrect'


user_name = input()
print(check_user(user_name))
```
